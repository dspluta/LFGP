```python
import numpy as np
import pickle
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from tqdm import tqdm as tqdm
```

```python
F_sample_odor_A = []
loading_sample_odor_A = []
mse_history_odor_A = []
for i in tqdm(range(10)):
    with open('results_odor_A_epoch_{}.pkl'.format(i + 1), 'rb') as fp:
        results = pickle.load(fp)
        F_sample, loading_sample, variance_sample, trace_sample, mse_history = results
        F_sample_odor_A.append(F_sample)
        loading_sample_odor_A.append(loading_sample)
        mse_history_odor_A.append(mse_history)
```

```python
fig = plt.figure(figsize=(4, 3), dpi=150)
for i in range(10):
    plt.plot(mse_history_odor_A[i], alpha=0.5, color='darkblue')
plt.title('MSE for Odor A Epochs')
plt.show()
```

```python
loading_sample = loading_sample_odor_A[0]

fig = plt.figure(figsize=(4, 3), dpi=150)
for i in range(100):
    plt.plot(loading_sample[i].flatten(), alpha=0.1, color='green')
plt.title('Regression Coefficient Posterior')
plt.show()
```

```python
fig = plt.figure(figsize=(4, 3), dpi=150)
for i in range(10):
    loading_sample = loading_sample_odor_A[i]
    test = np.stack(loading_sample).reshape((100, 30))
    posterior_median = np.median(test, axis=0)
    plt.plot(posterior_median, alpha=0.5)
plt.title('Median Loadings for Odor A Epochs')
plt.show()
```

```python
F_sample = F_sample_odor_A[1]

fig = plt.figure(figsize=(4, 3), dpi=150)
for i in range(100):
    plt.plot(F_sample[i][:, 0], alpha=0.1, color='orange')
for i in range(100):
    plt.plot(F_sample[i][:, 1], alpha=0.1, color='red')
plt.title('GP Posterior')
plt.show()
```

```python
fig = plt.figure(figsize=(4, 3), dpi=150)
for i in range(10):
    F_sample = F_sample_odor_A[i]
    test = np.stack(F_sample)
    gp1_posterior_median = np.median(test[:, :, 0], axis=0)
    gp2_posterior_median = np.median(test[:, :, 1], axis=0)
    plt.plot(gp1_posterior_median, alpha=0.5, color='orange')
    plt.plot(gp2_posterior_median, alpha=0.5, color='red')
plt.title('Median Factors for Odor A Epochs')
plt.show()
```
