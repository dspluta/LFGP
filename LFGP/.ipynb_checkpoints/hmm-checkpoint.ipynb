```python
from scipy.linalg import expm, logm
```

### Data generation

```python
def create_symmetric_matrix(upper, dim=5):
    """
    Create a symmetric matrix with elements in the upper triangle.
    
    Args
        upper: (numpy array) vector with concatenated rows in the upper triangle
        dim: (int) square matrix dimension
    """
    mat = np.zeros((dim, dim))
    current = 0
    for i in range(dim):  # fill in the upper triangle
        end = current + dim - i
        mat[i, i:] = upper[current:end]
        current = end
    for i in range(1, dim):  # fill in the lower triangle
        for j in range(i):
            mat[i, j] = mat[j, i]
    return mat
```

```python
F = np.zeros((1200, 2))
F[:, 0] = y1
F[:, 1] = y2
```

```python
B = np.random.normal(0, 0.5, [2, 15])
```

```python
log_series = np.matmul(F, B)
print(log_series.shape)
```

```python
for j in range(15):
    plt.plot(log_series[:1000, j])
plt.show()
```

```python
cov_series = np.zeros((1200, 5, 5))
for i in range(1200):
    cov_series[i, :, :] = expm(create_symmetric_matrix(log_series[i, :]))
time_series = np.zeros((1200, 5))
for i in range(1200):
    time_series[i, :] = np.random.multivariate_normal(np.zeros(5), cov_series[i, :, :], 1)
```

```python
for j in range(5):
    plt.plot(time_series[:1000, j])
plt.show()
```

```python
X = time_series[:1000, :]
print(X.shape)
```

### Hidden markov model

```python
from hmmlearn import hmm
```

```python
model = hmm.GaussianHMM(n_components=4, covariance_type='full', n_iter=100, algorithm='map').fit(X)
```

```python
Z = model.predict(X)
```

```python
plt.plot(Z)
plt.show()
```

```python
print(model.covars_)
```

```python
def predict_cov(Z, covs):
    """
    Predict covariance covariance Log-Euclidean vector time series.
    
    Args
        Z: (numpy array) predicted discrete latent states of length t
        covs: (numpy array) [k, n, n] covariance matrices for the k latent states
    """
    t = Z.shape[0]
    n = covs[0].shape[0]
    #cov_hat = np.zeros((t, n, n))
    log_hat = np.zeros((t, int(0.5 * n * (n + 1))))
    for i in range(t):
        latent = Z[i]
        #cov_hat[i, :, :] = covs[latent, :, :]
        log_hat[i, :] = logm(covs[latent, :, :])[np.triu_indices(n)]
    return log_hat
```

```python
log_hat = predict_cov(Z, model.covars_)
```
